```python
import numpy, matplotlib, pandas

csv = './openpowerlifting-2020-09-06.csv'
df = pandas.read_csv(csv,index_col=0,low_memory=False)
```

```python
df.head(3)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
# dropnaで欠損値が１つでも含まれるのを省く
data = df[['Sex','BodyweightKg', 'Best3SquatKg', 'Best3DeadliftKg', 'Best3BenchKg','Age','Country','Equipment']].dropna(how='any')
data.head(5)
```

```python
# df['Sex'] = df['Sex'].map({'M': 1, 'F': 0, 'Mx': -1})
# df

from sklearn.preprocessing import LabelEncoder

print(set(data.Sex))
print(set(data.Country))
print(set(data.Equipment))
```

```python
le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
# data['Country'] = le.fit_transform(data['Country'])
# data
```

```python
scatter_matrix(data)
plt.show()
```

```python
# 相関係数の総当たり
data.corr()
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(10,7))
data['Sex'].value_counts().plot(kind='bar')
plt.title('Gender ratio',fontsize=20)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

```python
# なんかかっこよくグラフ吐いてくれるライブラリ
import seaborn
```
